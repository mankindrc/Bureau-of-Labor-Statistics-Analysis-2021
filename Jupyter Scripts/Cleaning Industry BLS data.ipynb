```python
import pandas as pd
import numpy as np
import csv
```

<h4>For Industry data in particular, I only used specific occupations.<h4>

```python
df = pd.read_csv("peoutlook.csv")
```

<h4>drop unwanted columns<h4>

```python
df.drop(df.columns[[2,11,12]],axis=1,inplace = True)
```

<h4>Individual CSVs from the BLS do not specifiy occupation. Add column to each CSV for occupation<h4>

```python
df['Occupation'] = 'Pediatricians General'
```

<h4>For PostgreSQL's import function, BLS data needs to have all commas, asterisks, pounds, and double asterisks replaced from original source<h4>

```python
df['OCC_CODE'] = df['OCC_CODE'].replace(',', '',regex=True)
df['OCC_TITLE'] = df['OCC_TITLE'].replace(',', '',regex=True)
df['TOT_EMP'] = df['TOT_EMP'].replace(',', '',regex=True)
df['H_MEAN'] = df['H_MEAN'].replace(',', '',regex=True)
df['A_MEAN'] = df['A_MEAN'].replace(',', '',regex=True)
df['H_MEDIAN'] = df['H_MEDIAN'].replace(',', '',regex=True)
df['A_MEDIAN'] = df['A_MEDIAN'].replace(',', '',regex=True)
df['Date'] = df['Date'].replace(',', '',regex=True)
```

```python
df['OCC_CODE'] = df['OCC_CODE'].replace('*', '')
df['OCC_TITLE'] = df['OCC_TITLE'].replace('*', '')
df['TOT_EMP'] = df['TOT_EMP'].replace('*', '')
df['H_MEAN'] = df['H_MEAN'].replace('*', '')
df['A_MEAN'] = df['A_MEAN'].replace('*', '')
df['H_MEDIAN'] = df['H_MEDIAN'].replace('*', '')
df['A_MEDIAN'] = df['A_MEDIAN'].replace('*', '')
df['Date'] = df['Date'].replace('*', '')
```

```python
df['OCC_CODE'] = df['OCC_CODE'].replace('#', '')
df['OCC_TITLE'] = df['OCC_TITLE'].replace('#', '')
df['TOT_EMP'] = df['TOT_EMP'].replace('#', '')
df['H_MEAN'] = df['H_MEAN'].replace('#', '')
df['A_MEAN'] = df['A_MEAN'].replace('#', '')
df['H_MEDIAN'] = df['H_MEDIAN'].replace('#', '')
df['A_MEDIAN'] = df['A_MEDIAN'].replace('#', '')
df['Date'] = df['Date'].replace('#', '')
```

```python
df['OCC_CODE'] = df['OCC_CODE'].replace('**', '')
df['OCC_TITLE'] = df['OCC_TITLE'].replace('**', '')
df['TOT_EMP'] = df['TOT_EMP'].replace('**', '')
df['H_MEAN'] = df['H_MEAN'].replace('**', '')
df['A_MEAN'] = df['A_MEAN'].replace('**', '')
df['H_MEDIAN'] = df['H_MEDIAN'].replace('**', '')
df['A_MEDIAN'] = df['A_MEDIAN'].replace('**', '')
df['Date'] = df['Date'].replace('**', '')
```

```python
df.to_csv(r'cleanpeoutlook.csv', index = False)
```
